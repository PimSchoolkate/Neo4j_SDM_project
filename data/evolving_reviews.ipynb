```python
import os
import pandas as pd
import random
from lorem_text import lorem
import numpy as np
```

```python
direc = os.getcwd()
processed_data_path = os.path.join(direc,'processed')
update_data_path = os.path.join(direc,'evolving')
```

```python
def generate_extra_review_data(df):
    decision_list = []
    review_text_list = []
    for reviewers in df['reviewed_by']:
        text_length = random.randint(3,20)
        if '|' in str(reviewers):
            amount = len(reviewers.split('|'))
            decisions = []
            for dec in range(amount):
                if random.randint(0,100) < 20:
                    decisions.append('no')
                else:
                    decisions.append('yes')
            decision_list.append('|'.join(decisions))
        else:
            if random.randint(0,100) < 20:
                decision_list.append('no')
            else:
                decision_list.append('yes')
        review_text_list.append(lorem.words(text_length))
    df['decisions'] = decision_list  
    df['review_text'] = review_text_list
    return df 

def generate_org_data(df, author_identifier):
    organisations = ['The Useless Duck Company', 'UPC', 'Pies For Nodes Inc.', 'Free University Brussels', 'Twente University', 'Zelensky Inc.', 'Vueling', 'Lenovo', 'Acer', 'Gadjah Mada University']
    author_org = {}
    org_list = []
    for author in np.unique(df[author_identifier].astype(str)):
            author_org[str(author)] = random.choice(organisations)
    for key, value in df[author_identifier].astype(str).iteritems():
       org_list.append(author_org[str(value)])
    df['organisation'] = org_list
    return df
```

```python
article_data = pd.read_csv(os.path.join(processed_data_path, 'article.csv'), sep = ';')
proceeding_papers = pd.read_csv(os.path.join(processed_data_path, 'proceeding_papers.csv'), sep = ';')
article_data_update = article_data[['article', 'corresponding', 'reviewed_by']]
proceeding_papers_data_update = proceeding_papers[['inproceedings', 'corresponding', 'reviewed_by']]
article_data_update = generate_extra_review_data(article_data_update)
article_data_update = generate_org_data(article_data_update, 'corresponding')
proceeding_papers_data_update = generate_extra_review_data(proceeding_papers_data_update)
proceeding_papers_data_update = generate_org_data(proceeding_papers_data_update, 'corresponding')
article_data_update.to_csv(os.path.join(update_data_path, 'article_update.csv'), sep = ';', index=None)
proceeding_papers_data_update.to_csv(os.path.join(update_data_path, 'proceeding_papers_update.csv'), sep = ';', index=None)
```
